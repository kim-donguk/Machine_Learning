```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm


# Our dataset and targets
# (16,2) 배열 생성
X = np.c_[(.4, -.7),
          (-1.5, -1),
          (-1.4, -.9),
          (-1.3, -1.2),
          (-1.1, -.2),
          (-1.2, -.4),
          (-.5, 1.2),
          (-1.5, 2.1),
          (1, 1),
          # --
          (1.3, .8),
          (1.2, .5),
          (.2, -2),
          (.5, -2.4),
          (.2, -2.3),
          (0, -2.7),
          (1.3, 2.1)].T
    
# 리스트 Y 생성 0:8개 1:8개
Y = [0] * 8 + [1] * 8

# figure number
fignum = 1

# fit the model
# Kernel 값을 linear, poly, rbf 타입으로 각각 출력
# linear = 선형
# poly = 다항식 커널
# rbf = 가우시안 rbf 커널
for kernel in ('linear', 'poly', 'rbf'):

    # Support Vector Classifier, Kernel = linear, poly, rbf
    # gamma = 1
    #모델 생성
    clf = svm.SVC(kernel=kernel, gamma=1, C = 1000)
    clf.fit(X, Y)

    # plot the line, the points, and the nearest vectors to the plane
    # 선, 점 및 가장 가까운 벡터를 평면에 표시
    # figure 객체 생성
    plt.figure(fignum, figsize=(4, 3))
    # clear 함수
    plt.clf()

    # 산포 그래프 
    # 지지벡터(Support Vecotr) 표현
    # s = 마커의 크기, c = 색상, zorder = layer의 위치 
    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80,
                facecolors='none', zorder=10, edgecolors='k')
    plt.scatter(X[:8, 0], X[:8, 1], c='red', zorder=10, cmap=plt.cm.Paired,
                edgecolors='k')
    plt.scatter(X[8:16, 0], X[8:16, 1], c='blue', zorder=10, cmap=plt.cm.Paired,
                edgecolors='k')
    
    # tight 옵션은 모든 데이터를 볼 수 있을 정도로 축의 범위를 크게 넓히는 것
    plt.axis('tight')
    x_min = -3
    x_max = 3
    y_min = -3
    y_max = 3

    # Broadcasting 을 위한 차원 뻥튀기 shape = (200, 200)
    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    
    # ravel은 1차원 배열로 바꿔주는 함수
    Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(XX.shape)
    # figure 객체 생성
    plt.figure(fignum, figsize=(4, 3))
    plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
    # 분류선 설정
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                levels=[-.5, 0, .5])
    
    # x축 y축 지정
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)


    plt.xticks(())
    plt.yticks(())
    fignum = fignum + 1
plt.show()

# Gamma 값의 영향
# => gamma란 카우시안 커널의 폭을 제어하는 매개변수
# => 결정경계 주변의 샘플들이 결정경게에 영향을 크게주면서 구불구불한 경계가 만들어짐
# => 테스트 결과 Gamma 는 kernel이 rbf일 때 많은 영향을 줌

# C 값의 영향
# => c는 각 포인트의 중요도를 제한
# => c값이 커질수록 포인트들이 모델에 큰 영향을 주어 결정 경계를 정확하게 분류함
# => c는 overfitting과 under-fiiting 정도를 구분해줌
# => 테스트 결과 C는 kernel이 lienar일 때 많은 영향을 줌
```
